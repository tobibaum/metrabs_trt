```python
import numpy as np
from pandas import DataFrame
from glob import glob

files = glob('comparison/*.txt')

dat = []
for f in files:
    try:
        with open(f) as fh:
            lines = [l.strip() for l in fh.readlines()]
        d = {}

        name_parts = f.split('/')[1].split('.')[0].split('_')
        if len(name_parts) == 5:
            bbone = name_parts[1] + '_360'
        else:
            bbone = name_parts[1]
        det = name_parts[2]
        bs = int(name_parts[-1].split('=')[1])
        d['bbone'] = bbone
        d['det'] = det
        d['bs'] = bs

        ts = []
        for l in lines:
            try:
                t = float(l)
                ts.append(t)

            except ValueError:
                ps = l.split(' ')
                d[ps[0][:-1]] = float(ps[1][:-1])


        d['tfirst'] = ts[0]    
        d['tmean'] = np.mean(ts[1:])
        d['t/sample'] = np.mean(ts[1:]) / bs
        d['ts'] = ts
        d['f'] = f

        dat.append(d)
    except Exception as e:
        print('ERROR', e)
        print(f)
        print('\t', l)
        
df = DataFrame(dat)
df.sort_values(['bbone', 'det', 'bs'])
```
